```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('db/personal_transactions.csv')
df
```

```python
df = df.drop("account Name", axis = 1)
```

```python
df = df.drop("transaction Type", axis = 1)
df.head()
```

```python
df.sample(5)
```

```python
df['category'].unique()
```

```python
df
```

```python
# Calculate 60% of the dataset length
n_rows = len(df)
n_income = int(0.6 * n_rows)

# Randomly select 40% of rows to be "Income"
income_indices = df.sample(n=n_income, random_state=42).index

# Set those selected rows to "Income"
df.loc[income_indices, "type"] = "income"

# The rest remain as "Expense" (if needed, you can explicitly set them)
df.loc[~df.index.isin(income_indices), "type"] = "expense"
```

```python
df['type'].count()
```

```python
df['type'].value_counts()
```

```python
df['category'].unique()
```

```python
category_map = {
    'Shopping': 'Shopping & Retail',
    'Electronics & Software': 'Shopping & Retail',
    
    'Mortgage & Rent': 'Housing',
    'Home Improvement': 'Housing',
    'Utilities': 'Housing',
    'Internet': 'Housing',
    'Mobile Phone': 'Housing',
    'Television': 'Housing',
    
    'Restaurants': 'Food & Dining',
    'Fast Food': 'Food & Dining',
    'Coffee Shops': 'Food & Dining',
    'Groceries': 'Food & Dining',
    'Food & Dining': 'Food & Dining',
    
    'Gas & Fuel': 'Transportation',
    'Auto Insurance': 'Transportation',
    
    'Music': 'Entertainment',
    'Movies & DVDs': 'Entertainment',
    'Entertainment': 'Entertainment',
    
    'Haircut': 'Personal Care',
    'Alcohol & Bars': 'Personal Care',
    
    'Paycheck': 'Income',
    'Credit Card Payment': 'Debt Repayment'
}
```

```python
df['category'] = df['category'].map(category_map)
```

```python
df['category'].value_counts()
```

```python
# Save the updated dataset
df.to_csv("db/updated_dataset.csv", index=False)
```
